```python
import pandas as pd
import pickle
file = open('../data/players_stats.pkl', 'rb')
stats = pickle.load(file)
file.close()
```

```python
final_stats = pd.DataFrame()
for player in stats:
    name = player['info'].common_player_info.get_data_frame()['DISPLAY_FIRST_LAST'][0]
    data = player['career'].season_totals_regular_season.get_data_frame()
    data.insert(1, 'PLAYER_NAME', name)
    final_stats = final_stats.append(data)
```

```python
final_stats
```

```python
salary = pd.read_csv('../data/salary.csv')
```

```python
salary
```

```python
salary.columns = ['PLAYER_NAME', 'SEASON_ID', 'SALARY_NOMINAL', 'SALARY_REAL']
```

```python
final_df = final_stats.set_index(['PLAYER_NAME', 'SEASON_ID']).join(salary.set_index(['PLAYER_NAME', 'SEASON_ID']), how='left').dropna(how='all', subset=['SALARY_NOMINAL', 'SALARY_REAL'])
```

```python
final_df.to_csv('../data/stats.csv')
```
